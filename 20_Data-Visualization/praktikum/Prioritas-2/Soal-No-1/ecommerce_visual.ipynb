#### 1. Import Library

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2. Read Dataset JSON

```python
with open('ecommerce.json', 'r') as f:
    data = json.load(f)
```

#### 3. Konversi data menjadi DataFrame

```python
df = pd.DataFrame(data)
```

#### 4. Jumlah nilai transaksi per hari

```python
daily_transaction = df.groupby('transaction_date')['transaction_amount'].sum()
```

#### 5. Kategori barang yang paling banyak dibeli

```python
most_purchased_category = df['item_category'].value_counts().head(5)
```

#### 6. Jumlah nilai transaksi dengan metode pembayaran e-wallet

```python
ewallet_transaction = df[df['payment_method'] == 'e wallet']['transaction_amount'].sum()
```

#### 7. Nilai maksimal, minimal, dan tengah dari jumlah nilai transaksi

```python
max_transaction = df['transaction_amount'].max()
min_transaction = df['transaction_amount'].min()
median_transaction = df['transaction_amount'].median()
```

#### 8. Visualisasi Data

```python
plt.figure(figsize=(17, 10))
```

#### 9. Visualisasi Jumlah nilai transaksi per hari

```python
plt.figure(figsize=(10, 6))
daily_transaction.plot(kind='bar', color='skyblue')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Total Nilai Transaksi')
plt.xticks(rotation=45)

# Menyelaraskan layout subplot
plt.tight_layout()
plt.show()
```

#### 10. Visualisasi Kategori barang yang paling banyak dibeli

```python
plt.subplot(2, 2, 2)
most_purchased_category.plot(kind='bar', color='orange')
plt.title('Kategori Barang yang Paling Banyak Dibeli')
plt.xlabel('Kategori Barang')
plt.ylabel('Jumlah Pembelian')
```

#### 11. Visualisasi Jumlah nilai transaksi dengan metode pembayaran e-wallet

```python
plt.subplot(2, 2, 3)
plt.bar(['E-Wallet', 'Others'], [ewallet_transaction, df['transaction_amount'].sum() - ewallet_transaction], color=['blue', 'grey'])
plt.title('Jumlah Nilai Transaksi dengan Metode Pembayaran E-Wallet')
plt.ylabel('Total Nilai Transaksi')
```

#### 12. Visualisasi Nilai maksimal, minimal, dan tengah dari jumlah nilai transaksi

```python
plt.subplot(2, 2, 4)
plt.boxplot(df['transaction_amount'], vert=False)
plt.title('Distribusi Nilai Transaksi')
plt.xlabel('Total Nilai Transaksi')

# Tambahkan titik data untuk nilai maksimal, minimal, dan tengah
plt.scatter([max_transaction, min_transaction, median_transaction], [1, 1, 1], color='red', marker='o', label='Max, Min, Median')
plt.legend()

plt.tight_layout()
plt.show()
```
