#### 1. Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 2. Read Dataset

```python
df = pd.read_csv('survey.csv')
```

#### 3. Menghapus data duplikat dan data dengan atribut yang tidak lengkap

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

#### 4. Hapus baris-baris yang memiliki nilai "-" pada kolom favorite_front_end dan age

```python
df = df[df['favorite_front_end'] != '-']
df = df[df['age'] != '-']
```

#### 5. Agregasi data berdasarkan 'name', 'age'. dan 'name', 'salary_per_year_in_USD'

```python
aggregated_df = df.groupby(['name', 'age']).size().reset_index(name='count')
aggregated_salary_df = df.groupby(['name', 'salary_per_year_in_USD']).size().reset_index(name='count')
df
```

#### 6. Visualisasi Persentase responden berdasarkan umur

```python
plt.figure(figsize=(10, 6))
sns.histplot(aggregated_df['age'], bins=10, kde=True)
plt.title('Distribusi Umur Responden')
plt.xlabel('Umur')
plt.ylabel('Jumlah Responden')
plt.show()
```

#### 7. Visualisasi Bahasa pemrograman/framework yang paling banyak digunakan dalam pengembangan front end

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='favorite_front_end', data=df, order=df['favorite_front_end'].value_counts().index)
plt.title('Penggunaan Front End Framework / Bahasa Pemrograman')
plt.xlabel('Jumlah Responden')
plt.ylabel('Framework / Bahasa Pemrograman')
plt.show()
```

#### 8. Visualisasi Bahasa pemrograman/framework yang paling banyak digunakan dalam pengembangan back end

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='favorite_back_end', data=df, order=df['favorite_back_end'].value_counts().index)
plt.title('Penggunaan Back End Framework / Bahasa Pemrograman')
plt.xlabel('Jumlah Responden')
plt.ylabel('Framework / Bahasa Pemrograman')
plt.show()
```

#### 9. Persentase bahasa pemrograman/framework yang dipilih dalam pengembangan front end

```python
front_end_percentage = df['favorite_front_end'].value_counts() / len(df) * 100
plt.figure(figsize=(10, 6))
front_end_percentage.plot(kind='bar')
plt.title('Persentase Penggunaan Front End Framework / Bahasa Pemrograman')
plt.xlabel('Framework / Bahasa Pemrograman')
plt.ylabel('Persentase')
plt.show()
```

#### 10.  Rentangan gaji per tahun yang paling banyak berdasarkan hasil survei

```python
plt.figure(figsize=(10, 6))
sns.histplot(aggregated_salary_df['salary_per_year_in_USD'], bins=10, kde=True)
plt.title('Distribusi Gaji per Tahun')
plt.xlabel('Gaji per Tahun (USD)')
plt.ylabel('Jumlah Responden')
plt.show()
```
