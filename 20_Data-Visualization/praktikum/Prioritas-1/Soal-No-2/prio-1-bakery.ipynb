#### 1. Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### 2. Read Dataset

```python
df = pd.read_csv('bakery.csv')
df.set_index("transaction_id")
df
```

#### 3. Visualisasi jumlah nilai transaksi per hari

```python
transaction_date_result = df.groupby(["transaction_date"]).sum(numeric_only=True)

transaction_dates = transaction_date_result.index.get_level_values(0).tolist()
amount_total = transaction_date_result["transaction_amount"]

plt.figure(figsize=(15,7))

plt.title("Jumlah Nilai Transaksi per Hari")
plt.xlabel("Tanggal Transaksi")
plt.ylabel("Total Nilai Transaksi")

plt.plot(transaction_dates, amount_total, marker='o', linestyle='-', color='red')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

#### 4. Visualisasi jumlah roti yang paling banyak dipesan

```python
most_ordered_bread = df['item_name'].value_counts().idxmax()

print("Jenis roti yang paling banyak dipesan:", most_ordered_bread)

# Visualisasi jumlah roti yang paling banyak dipesan
plt.figure(figsize=(10, 6))
df['item_name'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Jumlah Roti Paling Banyak Dipesan')
plt.xlabel('Jenis Roti')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### 5. Visualisasi rata-rata kepuasan pelanggan per hari

```python
transaction_date_result = df.groupby(["transaction_date"]).mean(numeric_only=True)

transaction_dates = transaction_date_result.index.get_level_values(0).tolist()
cust_rating_averages = transaction_date_result["customer_rating"]

plt.figure(figsize=(10,6))

plt.title("Rata-Rata Kepuasan Pelanggan per Hari")
plt.xlabel("Tanggal")
plt.ylabel("Customer Rating")

plt.bar(transaction_dates, cust_rating_averages, color='orange')
plt.xticks(rotation=45)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
