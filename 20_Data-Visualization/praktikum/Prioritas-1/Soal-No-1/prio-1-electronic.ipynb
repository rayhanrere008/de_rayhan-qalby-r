#### 2. Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 3. Read Dataset

```python
df = pd.read_csv('electronic_shop.csv')
df.set_index("transaction_id")
df
```

#### 4. Line plot untuk jumlah nilai transaksi per hari (menggunakan matplotlib)

```python
transaction_date_result = df.groupby(["transaction_date"]).sum(numeric_only=True)

transaction_dates = transaction_date_result.index.get_level_values(0).tolist()
amount_total = transaction_date_result["transaction_amount"]

plt.figure(figsize=(15,7))

plt.title("Jumlah Nilai Transaksi per Hari")
plt.xlabel("Tanggal Transaksi")
plt.ylabel("Total Nilai Transaksi")

plt.plot(transaction_dates, amount_total, marker='o', linestyle='-')
plt.grid(True)

plt.show()
```

#### 5. Bar plot untuk jumlah customer per hari (menggunakan pandas)

```python
daily_cust = df.groupby(["transaction_date"]).nunique()["customer_name"]
daily_cust.plot(kind="bar", xlabel="Tanggal Transaksi", ylabel="Jumlah Customer", title="Jumlah Customer per Hari")
```

#### 6. Scatter plot untuk jumlah nilai transaksi per hari (menggunakan seaborn)

```python
sns.scatterplot(data=transaction_date_result, x='transaction_date', y='transaction_amount', marker='o')
plt.grid(True)
```
